```python
#Creo la lista viaggi, che funge da catalogo, successivamente includerà i dizionari che rappresenteranno ogni singolo viaggio.
viaggi=[]
```

```python
#Funzione per l'aggiunta di un nuovo viaggio.Non ha parametri. Con il ciclo for viene richiesto l'input di tutti gli elementi dei viaggi.
#e vengono inseriti nella nuova lista. Viene restituita la lista compilata.
def nuovo_viaggio(lista):
    new_viaggio = {
        "Nome_Viaggio": None,
        "Paese": None,
        "Tipologia": None,
        "Durata": None,
        "Livello_Sostenibilità": None
    }

    for elemento in new_viaggio:
        if elemento == "Durata":
            while True:
                try:
                    durata = int(input("Inserisci Durata (giorni): "))
                    if durata > 0:
                        new_viaggio["Durata"] = durata
                        break
                    else:
                        print("La durata deve essere positiva.")
                except ValueError:
                    print("Inserisci un numero valido.")

        elif elemento == "Livello_Sostenibilità":
            while True:
                try:
                    sostenibilità = int(input("Inserisci il livello di Sostenibilità da 0 a 5: "))
                    if 0 <= sostenibilità <= 5:
                        new_viaggio["Livello_Sostenibilità"] = sostenibilità
                        break
                    else:
                        print("Il livello di sostenibilità deve essere tra 0 e 5.")
                except ValueError:
                    print("Inserisci un numero valido.")

        else:
            while True:
                valore = input(f"Inserisci {elemento}: ").strip().lower()
                if valore:
                    new_viaggio[elemento] = valore
                    break
                else:
                    print("Input non valido.")

    for viaggio in lista:
        if (
            viaggio["Nome_Viaggio"] == new_viaggio["Nome_Viaggio"] and
            viaggio["Paese"] == new_viaggio["Paese"]
        ):
            print("Questo viaggio è già presente nel catalogo.")
            return None

    return new_viaggio
```

```python
#Funzione per visualizzare tutti i viaggi. Unico parametro è la lista che si desidera visualizzare. Semplice print
def visualizza_viaggi(lista):
    if not lista:
        print("Nessun viaggio registrato")
        return
    else:
        contatore = 1
        for viaggio in lista:
            print(f"\nViaggio {contatore}:")
            for chiave, valore in viaggio.items():
                print(f"  {chiave}: {valore}")
            contatore += 1
```

```python
#Funzione per ricercare un viaggio con il filtro del paese. Si chiede in input il Paese da filtrare per la ricerca.
#Con l'if si valuta la corrispondenza e, se trovata, si stampa in output. Se non si trova, si stampa un messaggio d'errore.
def ricerca_paese(lista):
    if not lista:
        print("Nessun viaggio registrato")
        return
    else:
        a=input("Inserisci il Paese del viaggio che cerchi: ").strip().lower()
        trovato=0
        for elemento in lista:
            if elemento["Paese"] == a:
                print(elemento)
                trovato+=1

        if trovato==0:
            print("Nessun viaggio trovato per questo paese. ")


#Funzione per ricercare un viaggio con il filtro della Tipologia. Si chiede in input la Tipologia da filtrare per la ricerca.
#Con l'if si valuta la corrispondenza e, se trovata, si stampa in output. Se non si trova, si stampa un messaggio d'errore.
def ricerca_tipologia(lista):
    if not lista:
        print("Nessun viaggio registrato")
        return
    else:
        a=input("Inserisci la tipologia di viaggio che cerchi: ").strip().lower()
        trovato=0
        for elemento in lista:
            if elemento["Tipologia"] == a:
                print(elemento)
                trovato+=1

        if trovato==0:
            print("Nessun viaggio trovato per questa tipologia. ")


#Funzione per ricercare un viaggio con il filtro della durata. Si chiede in input la durata del viaggio da filtrare per la ricerca.
#Con l'if si valuta la corrispondenza e, se trovata, si stampa in output. Se non si trova, si stampa un messaggio d'errore.
def ricerca_durata(lista):
    try:
        a = int(input("Inserisci la durata del viaggio che cerchi: "))
    except ValueError:
        print("Errore: inserisci un numero valido per la durata.")
        return

    trovato = 0
    for elemento in lista:
        if elemento["Durata"] == a:
            print(elemento)
            trovato += 1

    if trovato == 0:
        print("Nessun viaggio trovato con questa durata.")
```

```python
#Funzione per i filtri avanzati. Qua vengono richiesti 3 input diversi, Paese, Tipologia, Durata. Con un ciclo for e un contatore si cerca
#esattamente il viaggio che corrisponda a tutti e 3 gli input (condizionale con AND). E' previsto anche un messaggio d'errore nel caso non venga
#trovato nulla
def ricerca_filtri(lista):
    if not lista:
        print("Nessun viaggio registrato")
        return

    a = input("Inserisci il Paese del viaggio che cerchi: ").strip().lower()
    b = input("Inserisci la Tipologia del viaggio che cerchi: ").strip().lower()

    try:
        c = int(input("Inserisci la Durata del viaggio che cerchi: "))
    except ValueError:
        print("Errore: la durata deve essere un numero.")
        return

    trovato = 0
    for elemento in lista:
        if (
            elemento["Paese"].lower() == a and
            elemento["Tipologia"].lower() == b and
            elemento["Durata"] == c
        ):
            print(elemento)
            trovato += 1

    if trovato == 0:
        print("Nessun viaggio trovato con questi criteri.")
```

```python
#Funzione per statistiche: Numero di viaggi totali registrati. Parametro della lista dei viaggi e viene semplicemente stampato
#il numero totale di viaggi registrati, con la funzione len()
def totale_viaggi(lista):
    if not lista:
        print("Nessun viaggio registrato")
        return
    else:
        print("Attualmente ci sono ",len(lista), " viaggi registrati")


#Funzione per statistiche: Numero di viaggi per ogni singolo Paese. Parametro la lista dei viaggi. Tramite un ciclo for e una lista nuova
#vengono salvati e conteggiati tutti i paesi dei viaggi, inserendoli nella lista nuova.
def viaggi_per_paese(lista):
    if not lista:
        print("Nessun viaggio registrato")
        return
    else:
        a={}
        for elemento in lista:
            paese=elemento["Paese"]
            if paese in a:
                a[paese]+=1
            else:
                a[paese]=1

        print(a)

#Funzione per statistiche: Durata media dei viaggi registrati. Parametro lista dei viaggi. La media si calcola con somma/contatore di viaggi.
#Quindi dichiaro entrambe le variabili. Controllo che la lista non sia vuota e da li con un ciclo for aggiorno somma e contatore.
#Con il print finale stampo la media.
def durata_media(lista):
    if not lista:
        print("Nessun viaggio registrato")
        return
    else:
        somma=0
        a=0
        for elemento in lista:
            somma+=elemento["Durata"]
            a+=1

    if a>=1:
        print("La durata media di tutti i viaggi registrati è di ", somma/a, " giorni. ")
    else:
        print("Lista vuota")
```

```python
#Menù interattivo con tutti i comandi che richiamano le funzioni. Per ogni input si attiva una funzione.

while True:
    print("\n--- Menu Orizon ---")
    print("1. Aggiungi nuovo viaggio")
    print("2. Visualizza tutti i viaggi")
    print("3. Ricerca viaggi per Paese")
    print("4. Ricerca viaggi per Tipologia")
    print("5. Ricerca viaggi per Durata")
    print("6. Filtri avanzati")
    print("7. Statistiche")
    print("8. Esci")

    scelta = input("Scegli un'opzione: ")

    if scelta == "1":
        viaggio = nuovo_viaggio(viaggi)
        if viaggio:
            viaggi.append(viaggio)
    elif scelta == "2":
        visualizza_viaggi(viaggi)
    elif scelta == "3":
        ricerca_paese(viaggi)           
    elif scelta == "4":
        ricerca_tipologia(viaggi)    
    elif scelta == "5":
        ricerca_durata(viaggi)       
    elif scelta == "6":
        ricerca_filtri(viaggi)           
    elif scelta == "7":
        totale_viaggi(viaggi)
        viaggi_per_paese(viaggi)
        durata_media(viaggi)
    elif scelta == "8":
        print("Arrivederci!")
        break
    else:
        print("Scelta non valida, riprova.")
```
